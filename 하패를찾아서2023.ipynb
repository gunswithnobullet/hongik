```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains     #javascript 조작을 돕는 action chains
from bs4 import BeautifulSoup
import pandas as pd
import time
import warnings
warnings.filterwarnings(action='ignore')

from PIL import Image
import urllib.request as req
import openpyxl as xls
```

```python
driver = webdriver.Chrome('chromedriver.exe')
url = '''http://www.statiz.co.kr/stat.php?mid=stat&re=1&ys=2023&ye=2023&se=0&te=&tm=&ty=0&qu=all&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR&o2=OutCount&de=1&lr=0&tr=&cv=&ml=1&sn=30&pa=240&si=&cn='''
driver.get(url)
driver.maximize_window()
```

```python
frame0 = pd.DataFrame(columns = range(0,25))
```

```python
for num in range(3,20):
    xpath = '''//*[@id="fixcol"]/table/tbody/tr[{}]/td[2]/a'''.format(num)
    try:
        pitcher0 = driver.find_element_by_xpath(xpath)
        pitcher0.click()
        xpath1 = '''/html/body/div/div[1]/div/section[2]/div/div[1]/div/div[3]/div/div[2]/table/tbody/tr/td/a[8]'''
        pitcher1 = driver.find_element_by_xpath(xpath1)
        pitcher1.click()
        time.sleep(1)
        xpath2 = '''/html/body/div/div[1]/div/section[2]/div/div[2]/div/div[3]/div/a[3]'''
        pitcher2 = driver.find_element_by_xpath(xpath2)
        pitcher2.click()
        html = req.urlopen(driver.current_url)
        soup = BeautifulSoup(html, 'html.parser')
        content = soup.find_all('div', {'class' : 'col-xs-12 col-sm-4'})
        number = content[1].find_all('font')
        number1 = []
        for i in number:
            number1.append(i.get_text())
        driver.back()
        driver.back()
        driver.back()
        frame1 = pd.DataFrame(number1).transpose()
        frame0 = pd.concat([frame0, frame1], ignore_index=True)
    except Exception as e:
        print(f"예외 발생: {e}")
        pass
```
