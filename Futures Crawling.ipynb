```python
import urllib.request as req
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import openpyxl
from openpyxl import load_workbook

url = "https://www.koreabaseball.com/Futures/Schedule/BoxScore.aspx?leagueId=2&seriesId=0&gameId=20230917SKOB0&seasonId=2023"
html = req.urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
```

```python
awayplayerlist = []
awaypositionlist = []
awaypitcherlist = []
awaypitcherposlist = []

awaydiv_list = soup.find_all('div', {'class' : 'tbl-type03 tbl-futures03 data1 mb30'}) 
awayth_list = awaydiv_list[0].find_all('th', {'class' : 'name'})
for i in range(len(awayth_list)):
    awayplayerlist.append(awayth_list[i].text)   # 야수 이름
    
awaydiv_list2 = soup.find_all('div', {'class' : 'tbl-type03 tbl-futures03 data1 mb30'}) 
awayth_list2 = awaydiv_list2[0].find_all('th', {'class' : 'pos2'})
for i in range(len(awayth_list2)):
    awaypositionlist.append(awayth_list2[i].text)  # 야수 포지션    
    
awaypitcherdiv_list = soup.find_all('div', {'class' : 'tbl-type03 tbl-futures03 mb30'}) 
awaypitcherth_list = awaypitcherdiv_list[0].find_all('th', {'class' : 'name'})
for i in range(len(awaypitcherth_list)):
    awaypitcherlist.append(awaypitcherth_list[i].text)
    awaypitcherposlist.append(awaypitcherth_list[i].text)  # 투수 이름

for i in range(0,len(awaypitcherlist)):
    if i == 0:
        awaypitcherposlist[0] = '선발'
    else:
        awaypitcherposlist[i] = '구원'    # 투수 보직
```

```python
awayplayers_df = pd.DataFrame(awayplayerlist)
awaypositionlist_df = pd.DataFrame(awaypositionlist)
awayplayers = pd.concat([awaypositionlist_df,awayplayers_df],axis = 1).transpose()

awaypitcher_df = pd.DataFrame(awaypitcherlist).transpose()
awaypitcherpos_df = pd.DataFrame(awaypitcherposlist).transpose()
awaypitchers = pd.concat([awaypitcherpos_df, awaypitcher_df],axis = 0)

awaygame = pd.concat([awayplayers, awaypitchers], axis = 1)

awaygame
```

```python
path = r"C:\py_source\야구 크롤링\0917ganghwa.xlsx"

writer = pd.ExcelWriter(path, engine = 'xlsxwriter')
awaygame.to_excel(writer, index = False)
writer.save()
```
