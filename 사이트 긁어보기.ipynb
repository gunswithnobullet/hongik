```python
import requests
from bs4 import BeautifulSoup
```

```python
smallsoup = []
bigsoup = []

for i in range(0,27):
    url = f'https://chimhaha.net/stream_free?page={i+1}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text , 'html.parser')
    title = soup.find_all('span', {'class':'text'})
    if i+1 == 1:
        for j in range(5,len(title)):
            bigsoup.append(title[j].get_text())
    else:
        for k in range(3,len(title)):
            bigsoup.append(title[k].get_text())      
```

```python
bigsoup[-1]
```

```python
newsoup = []
search = '[침착맨] '
search2 = '[생중계] '

for i in range(0,len(bigsoup)):
    if search in bigsoup[i]:
        newsoup.append(bigsoup[i].strip(search))
    if search2 in bigsoup[i]: 
        newsoup.append(bigsoup[i].strip(search2))
```

```python
newsoup
```

```python
import pandas as pd

newsoup_df = pd.DataFrame(newsoup)
path = r"C:\py_source\투부 크롤링\침토이즘3.xlsx"

writer = pd.ExcelWriter(path, engine = 'xlsxwriter')
newsoup_df.to_excel(writer, index = False)
writer.save()
```

```python
cloudsoup = []

for i in range(0,len(newsoup)):
    splitsoup = newsoup[i].split(" ")
    for j in range(0,len(splitsoup)):
        cloudsoup.append(splitsoup[j])
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
new = soup.find_all('yt-formatted-string', {'class':'style-scope ytd-rich-grid-media'})
```

```python
newlist = []

for i in new:
    newlist.append(str(i))
```

```python
title_list = []
time_list = []
number_list = []

for i in range(len(newlist)):
    try:
        title = newlist[i].split('''id="video-title">''')[1].split("</yt-formatted-string>")[0]
        time = newlist[i].split("=")[1].split(''' 게시자:''')[1].split(" 침착맨 ")[1].split(" 전")[0]
        number = newlist[i].split("=")[1].split("게시자")[1].split("조회수 ")[1].split("회")[0]
        title_list.append(title)
        time_list.append(time)
        number_list.append(number)
    except:
        pass
```

```python
newlist[2].split('''id="video-title">''')[1]
```

```python
title_data = pd.DataFrame(title_list, columns = ['제목'])
time_data = pd.DataFrame(time_list, columns = ['날짜'])
number_data = pd.DataFrame(number_list, columns = ['조회수'])
data_all = pd.concat([title_data, time_data, number_data],axis = 1)
```
